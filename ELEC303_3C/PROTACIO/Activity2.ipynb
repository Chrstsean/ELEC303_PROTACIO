<h2>Which months record the highest number of serious accidents, and how does seasonality affect accident trends?</h2>
<h3>32. Serious accidents increase steadily from March to a peak in September, suggesting that higher traffic during summer months (holidays, tourism, and longer daylight hours) contributes to more accidents.</h3>
<h3>33. The lowest accident counts in February and December may reflect reduced travel activity, harsher winter weather discouraging driving, or improved caution by drivers during icy/wet conditions.</h3>

```python
highest_serious = accident[accident["Accident_Severity"] == "Serious"]["Month"].value_counts().sort_index()
```

```python
labels = ['Jan','Feb','Mar','April','May','Jun','Jul','Aug','Sept','Oct','Noc','Dec']
highest_serious.plot(kind='line',
                   figsize=(10,8),
                   marker='o')
plt.xlabel('Months')
plt.ylabel('Number of accidents')
plt.title('Serious accident per month')
plt.xticks(ticks=range(1, 13), labels=labels, rotation=60)
plt.show()
```
