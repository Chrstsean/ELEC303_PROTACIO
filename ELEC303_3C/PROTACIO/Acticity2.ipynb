<h1>DATA VISUALIZATION INSIGHTS</h1>

<h2>Monthly Road Accidents by year 2019 - 2022</h2>

<h2>23. 2019 consistently had the highest monthly accidents, peaking around October and November.</h2>

<h2>24. 2020 accident counts are lower than 2019, reflecting external disruptions, but they rise again in late months similar to pre-pandemic trends./h2>

<h2>25. 2022 shows the lowest accident counts overall, with a dramatic drop in December.</h2>

<h2>26. All years display a seasonal pattern, with accidents dipping in August and rising again towards October–November.</h2>

```python
monthly_data = accident.groupby(['Year','Month'])['Index'].count()
monthly_data = monthly_data.unstack(level=0)
monthly_data.plot(
    marker = 'o',
    figsize = (10,5),
    title = 'Monthly Road Accidents by Year 2019 - 2022'
)

plt.xticks(range(1,13), ['Jan','Feb','Mar','April','May','Jun','Jul','Aug','Sept','Oct','Noc','Dec'])
plt.grid(alpha=1)
plt.show()
```
